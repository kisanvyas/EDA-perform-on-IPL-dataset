### number of matches played in each season

```python
data['win_by']=np.where(data['win_by_runs']>0, 'Bat first', 'Bowl first')
data['win_by'].head()
```

### pie_chart for predict bowling first or batting first

```python
Win=data.win_by.value_counts()
labels=np.array(Win.index)
sizes=Win.values
colors = ['#FFBF00', '#FA8072']
plt.figure(figsize=(10,8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Match results', fontsize=15)
plt.axis('equal')
plt.show()
```

### Number of 4's heat by team

```python
four_data=complate_data[complate_data['batsman_runs']==4]
four_data.head()
```

```python
four_data.groupby('batting_team')['batsman_runs'].agg([('runs by four', 'sum'),('fours', 'count')])
```

### most fours hit by players 

```python
batsman_four=four_data.groupby('batsman')['batsman_runs'].agg([('four','count')]).reset_index().sort_values('four',ascending=0)
batsman_four.head()
```

```python
ax=batsman_four.iloc[:10,:].plot('batsman','four',kind='bar')
plt.title('most fours hit by players', fontsize=18)
plt.xticks(rotation=50)
plt.xlabel('player name', fontsize=15)
plt.ylabel("number of 4's", fontsize=15)
plt.show()
```

### Total number of 4's in each season

```python
ax=four_data.groupby('season')['batsman_runs'].agg([('four','count')]).reset_index().plot('season','four', kind='bar')
plt.title('total number of fours in each season', fontsize=18)
plt.xlabel('season',fontsize=15)
plt.ylabel("number of 4's",fontsize=15)
plt.show()
```

### Number of 6's hit by a team

```python
six_data = complate_data[complate_data['batsman_runs']==6]
six_data.head()
```

```python
six_data.groupby('batting_team')['batsman_runs'].agg([('run by six','sum'),('sixes','count')])
```

### most sixes hit by players

```python
batsman_six=six_data.groupby('batsman')['batsman_runs'].agg([('six','count')]).reset_index().sort_values('six',ascending=0)
batsman_six
```

```python
ax=batsman_six.iloc[:10,:].plot('batsman','six',kind='bar',color='gold')
plt.title('most sixes hit by players',fontsize=18)
plt.xlabel("Player's name", fontsize=15)
plt.ylabel('number of six', fontsize=15)
plt.show()
```

### total number of 6's in each season

```python
ax=six_data.groupby('season')['batsman_runs'].agg([('six','count')]).reset_index().plot('season','six',kind='bar',color='pink')
plt.title("total number of 6's in each season", fontsize=18)
plt.xlabel('season', fontsize=15)
plt.ylabel("number of 6's", fontsize=15)
plt.show()
```

### top 10 leading score 

```python
batsman_score=data_deli.groupby('batsman')['batsman_runs'].agg(['sum']).reset_index().sort_values('sum',ascending=False).reset_index(drop=True)
batsman_score
```

```python
batsman_score=batsman_score.rename(columns={'sum':'batsman_runs'})
print('TOP 10 LEADING SCORE OF PLAYER')
batsman_score.iloc[:10,:]
```

### most wickets taking by bowlers

```python
wicket_data = data_deli.dropna(subset=['dismissal_kind'])
wicket_data = wicket_data[~wicket_data['dismissal_kind'].isin(['run out','retired hurt','obstructing the feild'])]
```

```python
top_bowler=wicket_data.groupby('bowler')['dismissal_kind'].agg(['count']).reset_index().sort_values('count', ascending=False).reset_index(drop=True).iloc[:10,:]
top_bowler=top_bowler.rename(columns={'count':'wickets'})
top_bowler
```

### Dismissals in IPL
